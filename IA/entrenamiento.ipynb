```python
from tensorflow import keras
import numpy as np
import pandas as pd
import pyodbc
```

```python
def datosE():
    try:
        conexion = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=localhost;DATABASE=vigendb;Trusted_Connection=yes')
        cursor = conexion.cursor()
        cursor.execute("SELECT @@version;")
        row = cursor.fetchone()
        dataset=[]
        cursor.execute("SELECT * FROM poll")
        rows = cursor.fetchall()
        for row in rows:
            dataset.append(row[2:])
    except Exception as ex:
        print(ex)
    return np.array(dataset)
```

```python
train_data=np.transpose(datosE())[:-1]
train_data=np.transpose(train_data)
train_labels= np.transpose(datosE())[-1]
conjunto=set(train_labels)
print(conjunto)
```

```python

model = keras.Sequential([
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(len(conjunto), activation='softmax')
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(np.array(train_data), np.array(train_labels), epochs=500)
```

```python
model.save('model/myModel.h5')
model.save_weights('model/myWeights.h5')
```
